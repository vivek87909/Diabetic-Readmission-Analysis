```python

import pandas as pd

# 1. Check if 'age' column actually exists
if 'age' not in df.columns:
    print("ERROR: Column 'age' not found. Did you rename it to 'age_group' already?")
    print("Columns available:", df.columns.tolist())

else:
    # 2. Clean the 'age' column first (Remove extra spaces which cause errors)
    # This converts to string and strips whitespace
    df['age'] = df['age'].astype(str).str.strip()

    # 3. Check the first few values to see what we are dealing with
    print("Sample values in 'age':", df['age'].unique()[:5])

    # 4. Define the Dictionary
    age_mapping = {
        '[0-10)': '0-30 (Young)',
        '[10-20)': '0-30 (Young)',
        '[20-30)': '0-30 (Young)',
        '[30-40)': '30-60 (Adult)',
        '[40-50)': '30-60 (Adult)',
        '[50-60)': '30-60 (Adult)',
        '[60-70)': '60+ (Senior)',
        '[70-80)': '60+ (Senior)',
        '[80-90)': '60+ (Senior)',
        '[90-100)': '60+ (Senior)'
    }

    # 5. Apply Map (If value is not in dict, keep original)
    # We use a lambda function: "If x is in the map, use the map. Else, keep x".
    df['age_group'] = df['age'].apply(lambda x: age_mapping.get(x, 'Unknown'))

    # 6. Verify
    print("\nSUCCESS. New Age Group Distribution:")
    print(df['age_group'].value_counts())
```

```python
from scipy.stats import chi2_contingency, ttest_ind

print("--- STARTING STATISTICAL ANALYSIS ---")
```

```python
df['readmitted_num'] = df['readmitted_num'].apply(lambda x: 0 if x == 'No' else 1)
```

```python
numeric_cols = ['time_in_hospital','num_lab_procedures','num_procedures', 
                'num_medications', 'number_diagnoses', 'readmitted_num']
```

```python
plt.figure(figsize=(10,8))
corr = df[numeric_cols].corr()
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Corellation Matrix: Numeric Features vs Readmission")
plt.show()
```

```python
print("\n>>> TEST 1: Chi-Square (Gender vs Readmission)")
```

```python
contingency_table = pd.crosstab(df['gender'],df['readmitted'])
print(contingency_table)
```

```python
chi2 , p, dof, expected = chi2_contingency(contingency_table)
print(f"P-value: {p:5f}")
```

```python

if p < 0.05:
    print("CONCLUSION: Statistically Significant. Gender affects Readmission rates.")
else:
    print("CONCLUSION: Not Significant. Differences are due to chance.")
```

```python
print("\n>>> TEST 2: T-Test (Length of Stay)")
```

```python
group_readmitted = df[df['readmitted_num'] == 1]['time_in_hospital']
group_safe = df[df['readmitted_num'] == 0]['time_in_hospital']
```

```python
t_stat, p_val = ttest_ind(group_readmitted, group_safe)
```

```python
print(f"Avg LOS (Readmitted): {group_readmitted.mean():.2f} days")
print(f"Avg LOS (Not Readmitted): {group_safe.mean():.2f} days")
print(f"P-Value: {p_val:.5f}")
```

```python
if p_val < 0.05:
    print("CONCLUSION: Significant. Readmitted patients have a different LOS.")
else:
    print("CONCLUSION: Not Significant.")

```
